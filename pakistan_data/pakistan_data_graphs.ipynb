```python
import pandas as pd
```

```python
gdp_per_cap = pd.read_csv('GDP_per_capita.csv', skiprows=3)
```

```python
gdp_per_cap.drop(['Country Code', 'Indicator Name', 'Indicator Code'], inplace=True, axis=1)
```

```python
gdp_per_cap = gdp_per_cap[gdp_per_cap['Country Name'].isin(['India', 'Pakistan'])]
```

```python
gdp_per_cap
```

```python
gdp_per_cap = pd.melt(gdp_per_cap, id_vars=['Country Name'], var_name='Year', value_name='GDP_Per_Capita')
```

```python

gdp_per_cap
```

```python
gdp_per_cap = gdp_per_cap.pivot(index='Year', columns='Country Name', values='GDP_Per_Capita')
```

```python
gdp_per_cap
```

```python
gdp_per_cap.dropna(inplace=True)
```

```python
gdp_per_cap
```

```python
gdp_per_cap.to_csv('India_vs_Pakistan_GDP_per_capita.csv')
```
